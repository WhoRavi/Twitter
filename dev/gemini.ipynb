```python
import os
from google import genai

# 1. Setup the client
client = genai.Client(api_key=os.getenv("GEMINI_API_KEY"))

# 2. Define the prompt and model
model_id = "models/imagen-4.0-generate-001"
prompt = "Royal Enfield continental GT 650 in a scenic mountain road with retro style. Do not show any people in the image."

try:
    # 3. Generate the image
    print("Generating image...")
    result = client.models.generate_images(
        model=model_id,
        prompt=prompt,
        config=dict(
            number_of_images=1,
            output_mime_type="image/jpeg",
            aspect_ratio="16:9",
            image_size="1K",
        ),
    )

    # 4. Save the image
    if not result.generated_images:
        print("No images generated.")
    else:
        for n, generated_image in enumerate(result.generated_images):
            filename = f"gt_{n}.jpg"
            generated_image.image.save(filename)
            print(f"Success! Image saved as {filename}")

except Exception as e:
    print(f"An error occurred: {e}")
```
